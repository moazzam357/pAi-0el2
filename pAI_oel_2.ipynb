```python
import pandas as pd

df= pd.read_csv('StudentsPerformance.csv')

df.head()
```

```python
# Convert categorical columns to dummy variables
df_dummy = pd.get_dummies(df)

# Remove features with zero variance (constant)
from sklearn.feature_selection import VarianceThreshold
selector = VarianceThreshold()
df_reduced_array = selector.fit_transform(df_dummy) # Renamed variable for clarity

# Get selected column names
selected_cols = df_dummy.columns[selector.get_support()]

# Create a new DataFrame from the reduced array with the selected column names
df_processed = pd.DataFrame(df_reduced_array, columns=selected_cols, index=df_dummy.index)

df = df_processed


df
```

```python
# Fill missing numerical values with mean
for col in df.select_dtypes(include='number').columns:
    df[col].fillna(df[col].mean(), inplace=True)

# Fill missing categorical values with mode
for col in df.select_dtypes(include='object').columns:
    df[col].fillna(df[col].mode()[0], inplace=True)
```
